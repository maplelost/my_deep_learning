```python
import torch
from torch import nn
from d2l import torch as d2l

batch_dize = 256
train_iter , test_iter = d2l.load_data_fashion_mnist(batch_dize)
```

```python
# 展平层,将任何维度的tensor转换为二维的tensor,在线性层前调整网络输入的形状
# linear,线性层,输入为784,输出为10
# Sequential,扩凿器
net = nn.Sequential(nn.Flatten(), nn.Linear(784,10))

def init__weights(m):
    if type(m) == nn.Linear:
        nn.init.normal_(m.weight, std=0.01)   # 将m.weight初始化成均值为零,方差为0.01的函数
        
net.apply(init__weights)
```

```python
loss = nn.CrossEntropyLoss()
```

```python
trainer = torch.optim.SGD(net.parameters(),lr = 0.1)
```

```python
for x in net.parameters():
    print(x[0])
```

```python
num_epochs = 10
d2l.train_ch3(net, train_iter, test_iter, loss, num_epochs, trainer)
```
